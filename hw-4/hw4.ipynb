```python
import numpy as np
import math
import matplotlib.pylab as plt
```

# 1.1 Calc matrix

```python
a = np.array([[1,-4], [3, 5], [-1, 0]])
b = np.array([[1, 2, 7], [-3, -4, 0], [5, 2, 1]])
c = np.array([[6, -3, 9], [4, -5, 2], [8, 1, 5]])
# print("A:")
print("A:\n", a)
print("B:\n", b)
print("C:\n", c, "\n")

at = a.T
print("At:\n", at)

bt = b.T
print("Bt:\n", bt, "\n")

d = at.dot(c) - 2*at.dot(bt)
print("Result:\n", d)
```

# 1.4 SVD and Approximation

```python
size = 100
# a = np.random.uniform(low=0.0, high=1000.0, size=(size, size))
a=np.random.randint(0, 1000, size=(size,size))
print("Init matrix A:")
print(a, "\n")

u,s,v = np.linalg.svd(a, compute_uv=True)
print("SVD")
print('u:\n',u)
print('s:\n',s)
print('v:\n',v, "\n")

e = np.eye(size)
s = s*e

print("Approximation:")
err_dict = {}
for r in range(2, size):
    a_appr = (u[:, :r].dot(s[:r, :r])).dot(v.T[:r, :])
#     print('a_appr:\n', a_appr, "\n")
    
    err = np.linalg.norm(a - a_appr)
    print("r =", r, "=> error =", err)
    err_dict[r] = err

plt.plot(*zip(*sorted(err_dict.items())))
plt.show()

```
